```python
import pandas as pd

ALL_TOKENS = pd.read_csv('Data\CORPUS\FULL_CORPUS.csv')
```

```python
VOCAB = ALL_TOKENS.term_str.value_counts().to_frame('n')
VOCAB.index.name = 'term_str'
```

```python
import numpy as np

VOCAB = ALL_TOKENS.term_str.value_counts().to_frame('n')
VOCAB.index.name = 'term_str'
VOCAB['p'] = VOCAB.n / VOCAB.n.sum()
VOCAB['i'] = -np.log2(VOCAB.p)
VOCAB['n_chars'] = VOCAB.index.str.len()
VOCAB
```

```python
VOCAB['max_pos_group'] = ALL_TOKENS[['term_str','pos_group']].value_counts().unstack(fill_value=0).idxmax(1)
VOCAB['max_pos'] = ALL_TOKENS[['term_str','pos']].value_counts().unstack(fill_value=0).idxmax(1)
VOCAB
```

```python
TPM1 = ALL_TOKENS[['term_str','pos_group']].value_counts().unstack()
TPM1
VOCAB['n_pos_group'] = TPM1.count(1)
TPM2 = ALL_TOKENS[['term_str','pos']].value_counts().unstack()
VOCAB['n_pos'] = TPM2.count(1)
```

```python
import nltk
nltk.download('stopwords')
sw = pd.DataFrame({'stop': 1}, index=nltk.corpus.stopwords.words('english'))
sw.index.name='term_str'
sw.head()
if 'stop' not in VOCAB.columns:
    VOCAB = VOCAB.join(sw)
    VOCAB['stop'] = VOCAB['stop'].fillna(0).astype('int')
VOCAB
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.lancaster import LancasterStemmer

p_stemmer = PorterStemmer()
s_stemmer = SnowballStemmer(ignore_stopwords=True, language='english')
l_stemmer = LancasterStemmer()
# VOCAB['p_stem'] = VOCAB.apply(lambda x: stemmer.stem(x.name), 1)
# VOCAB['p_stem'] = VOCAB.reset_index().term_str.apply(stemmer.stem)
VOCAB['porter_stem'] = [p_stemmer.stem(term_str) for term_str in VOCAB.index]
VOCAB['snowball_stem'] = [s_stemmer.stem(term_str) for term_str in VOCAB.index]
VOCAB['lancaster_stem'] = [l_stemmer.stem(term_str) for term_str in VOCAB.index]
VOCAB
```

```python
VOCAB.to_csv('Data\VOCAB\FULL_VOCAB.csv')
```
