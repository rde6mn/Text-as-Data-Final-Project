## Source Document Structure

The OHCO structure varies based on the novel in the corpus. For The Mysteries of Udolpho, it is ['vol_num', 'chap_num', 'para_num', 'sent_num', 'token_num'] ,for THe Old ENglish Baron, it is ['para_num', 'sent_num', 'token_num'], for the Castle of Otranto, it is ['chap_num', 'para_num', 'sent_num', 'token_num'], for Melmoth the Wanderer, it is ['chap_num', 'para_num', 'sent_num', 'token_num']. **For clarity however, each text will be formatted so that it follows the structure ['chap_num', 'para_num', 'sent_num', 'token_num']**. For works without chapters, everything will be put under a chap_num 1, and for works with volumes, only the first of the volumes will be investigated. THis ensures that there is a standard OHCO across all texts, which allows us to combine them into corpuses with ease.



The final OHCO structure is ['chap_num', 'para_num', 'sent_num', 'token_num'] across the CORPUS's for each era.
