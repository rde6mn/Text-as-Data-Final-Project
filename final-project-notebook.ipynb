## Source Document Structure

The OHCO structure varies based on the novel in the corpus. For The Mysteries of Udolpho, it is ['vol_num', 'chap_num', 'para_num', 'sent_num', 'token_num'] ,for THe Old ENglish Baron, it is ['para_num', 'sent_num', 'token_num'], for the Castle of Otranto, it is ['chap_num', 'para_num', 'sent_num', 'token_num'], for Melmoth the Wanderer, it is ['chap_num', 'para_num', 'sent_num', 'token_num']. **For clarity however, each text will be formatted so that it follows the structure ['chap_num', 'para_num', 'sent_num', 'token_num']**. For works without chapters, everything will be put under a chap_num 1, and for works with volumes, only the first of the volumes will be investigated. THis ensures that there is a standard OHCO across all texts, which allows us to combine them into corpuses with ease.



The final OHCO structure is ['chap_num', 'para_num', 'sent_num', 'token_num'] across the CORPUS's for each era.

# Interpretation

Going into this project, I wanted to see if Gothic novels were different across the features that I had in my LIB (era, date, and novel_type). In basically all the areas that I investigated, it seemed like across all the 3 descriptive features, only one category would be different from the rest. For the feature ‘era’, I investigated it in riff 2 with a sentiment plot across the 7 main emotions from the salex_nrc. Here, I found that of the 4 different eras, only the high and classic eras differed from the others, and those differences occurred for different emotions. I believe this is because those two were the earliest eras, and as the eras stabilized, the overall sentiments stabilized for gothic novels. I explored the date feature in riff 3 with a plot where I looked at the vector distances of books in a plot where the axes were represented by “demon” and “church”. The books from dates before 1850 were spread throughout the plot, but the vector distances were smaller for books after 1850, indicating that they were more similar to each other. I explored the feature novel_type the most because date and era were very related, so observations across either feature would most likely have pretty high correlation with each other. The novel_type feature was explored in the PCA plots, the LDA +PCA plot, and riff 1 where I looked at the tSNE plot for just romance type novels. In both PCA plots, the thriller novel type was different from both romance and horror across both PC0 and PC1. Thriller was also different from the other two for PC0 in the LDA plot. For the tSNE plot for romance, the plot was vastly different from the plot for all the genres combined (full dataset), signifying that romance differs from the overall trend. My main takeaways are that across time, the overall emotions and vibes of the songs stabilize, as shown with the features date and era. As time progressed, the sentiment values were similar for the more modern eras, and the vector distances were not as great across the words ‘demon’ and ‘church’. For the novel type conclusions, thriller was vastly different from both romance and horror, but I think that sub-genres may have been at work in making both romance and horror similar to each other. The romance novels may have had horror as a subgenre and vice versa, which would have messed up the patterns that I was observing. For future work, I would increase the dataset size to see if data and era related patterns persist even when the corpus is more fleshed out, and I would figure out a way to deal with books that have multiple genres. Overall, while there weren’t major differences between all levels for each of my features, there were distinctions between at least one level from the rest.
