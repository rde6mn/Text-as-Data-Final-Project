```python
import pandas as pd

ALL_TOKENS = pd.read_csv('Data\CORPUS\FULL_CORPUS.csv')
```

# VOCAB

```python
VOCAB = ALL_TOKENS.term_str.value_counts().to_frame('n')
VOCAB.index.name = 'term_str'
```

```python
import numpy as np

VOCAB = ALL_TOKENS.term_str.value_counts().to_frame('n')
VOCAB.index.name = 'term_str'
VOCAB['p'] = VOCAB.n / VOCAB.n.sum()
VOCAB['i'] = -np.log2(VOCAB.p)
VOCAB['n_chars'] = VOCAB.index.str.len()
VOCAB
```

```python
VOCAB['max_pos_group'] = ALL_TOKENS[['term_str','pos_group']].value_counts().unstack(fill_value=0).idxmax(1)
VOCAB['max_pos'] = ALL_TOKENS[['term_str','pos']].value_counts().unstack(fill_value=0).idxmax(1)
VOCAB
```

```python
TPM1 = ALL_TOKENS[['term_str','pos_group']].value_counts().unstack()
TPM1
VOCAB['n_pos_group'] = TPM1.count(1)
TPM2 = ALL_TOKENS[['term_str','pos']].value_counts().unstack()
VOCAB['n_pos'] = TPM2.count(1)
```

```python
import nltk
nltk.download('stopwords')
sw = pd.DataFrame({'stop': 1}, index=nltk.corpus.stopwords.words('english'))
sw.index.name='term_str'
sw.head()
if 'stop' not in VOCAB.columns:
    VOCAB = VOCAB.join(sw)
    VOCAB['stop'] = VOCAB['stop'].fillna(0).astype('int')
VOCAB
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.lancaster import LancasterStemmer

p_stemmer = PorterStemmer()
s_stemmer = SnowballStemmer(ignore_stopwords=True, language='english')
l_stemmer = LancasterStemmer()
# VOCAB['p_stem'] = VOCAB.apply(lambda x: stemmer.stem(x.name), 1)
# VOCAB['p_stem'] = VOCAB.reset_index().term_str.apply(stemmer.stem)
VOCAB['porter_stem'] = [p_stemmer.stem(term_str) for term_str in VOCAB.index]
VOCAB['snowball_stem'] = [s_stemmer.stem(term_str) for term_str in VOCAB.index]
VOCAB['lancaster_stem'] = [l_stemmer.stem(term_str) for term_str in VOCAB.index]
VOCAB
```

# TFIDF

```python
tf_method = 'double_norm'         # sum, max, log, double_norm, raw, binary
tf_norm_k = .5            # only used for double_norm
idf_method = 'standard'   # standard, max, smooth
gradient_cmap = 'YlGnBu'
```

```python
print('TF method:', tf_method)

if tf_method == 'sum':
    TF = DTCM.T / DTCM.T.sum()

elif tf_method == 'smooth':
    TF = (DTCM.T / DTCM.T.sum()) + 1

elif tf_method == 'max':
    TF = DTCM.T / DTCM.T.max()
    
elif tf_method == 'log':
    TF = np.log2(1 + DTCM.T)
    
elif tf_method == 'raw':
    TF = DTCM.T
    
elif tf_method == 'double_norm':
    TF = ((DTCM.T + .5) / (DTCM.T.max() + .5)) + .5
    
elif tf_method == 'binary':
    TF = DTCM.T.astype('bool').astype('int')
    
TF = TF.T
```

```python
DF = DTCM.astype('bool').sum() 
DF
```

```python
N = DTCM.shape[0]
```

```python
print('IDF method:', idf_method)

if idf_method == 'standard':
    IDF = np.log2(N / DF)

elif idf_method == 'max':
    IDF = np.log2(DF.max() / DF) 

elif idf_method == 'plus':
    IDF = np.log2(N / DF) + 1

# This is what SciKit Learn uses
elif idf_method == 'smooth':
    IDF = np.log2((1 + N) / (1 + DF)) + 1
```

```python
TFIDF = TF * IDF
TFIDF.head()
```

## TFIDF_L2

```python
def get_norm(X, p:int = 1):
    if p == 0:
        return X.astype(bool).astype(int)
    elif p > 0 and type(p) == int:
        unit_norm = (X**p).sum(1)**(1/p)
        return X.div(unit_norm, axis=0)
    else:
        raise ValueError("p must be an integer greater than 0")
    

TFIDF_L2 = get_norm(TFIDF, 2)
```
